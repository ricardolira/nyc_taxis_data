## Setup do projeto

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pytz
import gmaps
import descartes
import geopandas as gpd

from shapely.geometry import Point, Polygon
from random import randint
from dateutil.parser import parse
from datetime import datetime
%matplotlib inline
```

```python
ROOT = os.path.abspath("")
dataset_files = [
    os.path.join(ROOT, "data", filename) 
    for filename in os.listdir(os.path.join(ROOT, "data")) if filename.split('.')[-1] == 'json'
]
all_data = []
crs = {"init": "epsg:4326"}
for file in dataset_files:
    print(f"Lendo dados: {file}...")
    data_set = pd.read_json(file, lines=True)
    all_data.append(data_set)
print("FIM !!")
data_set = pd.concat(all_data)
```

## Transformando dados referentes às datas em *datetime objects*

```python

data_set['pickup_datetime'] = data_set['pickup_datetime'].apply(
    lambda datetime_str: parse(datetime_str)
)
data_set['dropoff_datetime'] = data_set['dropoff_datetime'].apply(
    lambda datetime_str: parse(datetime_str)
)
```

Para auxiliar nesse momento, faz-se necessário a criação de funções auxiliares que possam facilitar a extração de alguns parâmetros estatísticos de localização.

```python
def get_stats(column):
    mean = column.mean()
    dev = column.std()
    max_value = column.max()
    min_value = column.min()
    return mean, dev, max_value, min_value
```

```python
columns = ['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude']
pickup_longitude_stats = get_stats(data_set['pickup_longitude'])
pickup_latitude_stats = get_stats(data_set['pickup_latitude'])
dropoff_longitude_stats = get_stats(data_set['dropoff_longitude'])
dropoff_latitude_stats = get_stats(data_set['dropoff_latitude'])
```
